```python
from src.ising import IsingModel
import numpy as np
import matplotlib.pyplot as plt
```

```python
nsteps = 50
model = IsingModel(size=50, kb=1, J=1, H=0, n_mcsteps=nsteps)
temperatures, energies, magnetization = model.run_mc(1, 10)
```

```python
steps = range(2 * nsteps - 1)

plt.subplot(3, 1, 1)
plt.plot(steps, energies)

plt.subplot(3, 1, 2)
plt.plot(steps, magnetization)

plt.subplot(3, 1, 3)
plt.plot(steps, temperatures)
```
